###Exercício de Código | Intervalo de confiança e distribuição amostral

```python
#Adicionando as informações na variável:
#Distribuição amostral e intervalo de confiança

salarios = dados['SALARIO']
```

```python
salarios
```

##média amostral e o desvio amostral:

```python
salarios = dados["SALARIO"].dropna()

print("Quantidade:", len(salarios))
print("Média:", np.mean(salarios))
print("Desvio:", np.std(salarios))
print("Mínimo:", salarios.min())
print("Máximo:", salarios.max())
```

```python
media_amostral=np.mean(salarios)
media_amostral
```

```python
desvio_amostral = np.std(salarios)
desvio_amostral
```

##nível de confiança:



```python
nivel_confianca = 0.95
```

```python
#tamanho da amostra:
tamanho_amostra = len(salarios)
tamanho_amostra
```

```python
#erro padrão
from scipy import stats
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
#Intervalo de confiança
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra - 1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

###Exercício de Código | Featuring engineering

```python
#Criando a função | Featuring engineering

def preencher_nivel (gestor, nivel):
  if gestor==1:
    return "Pessoa gestora"
  else:
    return nivel
```

```python
#Aplicando a função

dados['NOVO NIVEL'] = dados.apply(lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']), axis=1)
```

```python
##Calcular variáveis discretas e correlação entre as tabelas cruzadas:
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = stats.chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape) - 1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dados["COR/RACA/ETNIA"], dados["NIVEL DE ENSINO"])
```
